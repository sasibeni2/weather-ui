### 📏 np.linspace - Egyenletesen elosztott pontok

A `np.linspace` **lineárisan elosztott** számokat generál egy intervallumban.

**Szintaxis:**
```python
np.linspace(start, stop, num, endpoint=True)
```

**Paraméterek:**
- `start`: Kezdőérték
- `stop`: Végérték
- `num`: Hány pontot generáljon (alapértelmezett: 50)
- `endpoint`: Tartalmazza-e a végpontot (alapértelmezett: True)

**Példák:**
```python
np.linspace(0, 10, 5)      # [0.0, 2.5, 5.0, 7.5, 10.0]
np.linspace(0, 1, 11)      # [0.0, 0.1, 0.2, ..., 1.0]
np.linspace(0, 10, 5, endpoint=False)  # [0.0, 2.0, 4.0, 6.0, 8.0]
```

**np.linspace vs np.arange:**

| np.linspace | np.arange |
|-------------|-----------|
| **Pontok száma** megadása | **Lépésköz** megadása |
| `linspace(0, 10, 100)` | `arange(0, 10, 0.1)` |
| Endpoint inkluzív | Endpoint exkluzív |
| Lebegőpontos számok | Int vagy float |
| Mindig pontosan N pont | Lépésköztől függ |

**Mikor használd linspace-t?**
- ✅ **Függvény plottolás**: Smooth görbékhez sok pont kell
- ✅ **Ismert pontszám**: Tudod hány mintát akarsz
- ✅ **Szimmetria**: Start és stop egyaránt benne van

**Mikor használd arange-t?**
- ✅ **Fix lépésköz**: "Minden 0.5-ös értéknél"
- ✅ **Egész számok**: `arange(0, 10)` → [0,1,2,...,9]

**Grafikon minőség:**
```python
# Rossz (durva görbe)
x = np.linspace(0, 10, 10)  # Csak 10 pont

# Jó (sima görbe)
x = np.linspace(0, 10, 100)  # 100 pont → szebb

# Túlzás (feleslegesen lassú)
x = np.linspace(0, 10, 10000)  # 10,000 pont
```

**💡 Tipp:** Általában 100-1000 pont elég sima görbékhez!

### 📐 Többpaneles Ábrák (Subplots)

A **subplots** lehetővé teszi több grafikon egyidejű megjelenítését egy ábrán!

**Alapszintaxis:**
```python
fig, ax = plt.subplots(nrows, ncols, figsize=(width, height))
```

**Paraméterek:**
- `nrows`: Sorok száma
- `ncols`: Oszlopok száma
- `figsize`: Ábra mérete (inch-ben!) → (szélesség, magasság)
- `fig`: Figure objektum (egész ábra)
- `ax`: Axes objektum (egyes panelek)

**Példák:**
```python
# 1 sor, 2 oszlop → két panel egymás mellett
fig, ax = plt.subplots(1, 2, figsize=(12, 5))
ax[0].plot(x, y1)  # Bal panel
ax[1].plot(x, y2)  # Jobb panel

# 2 sor, 2 oszlop → 2×2 grid
fig, ax = plt.subplots(2, 2, figsize=(10, 10))
ax[0, 0].plot(x, y1)  # Bal felső
ax[0, 1].plot(x, y2)  # Jobb felső
ax[1, 0].plot(x, y3)  # Bal alsó
ax[1, 1].plot(x, y4)  # Jobb alsó
```

**Mikor használd?**
- ✅ Összehasonlítások: Több modell eredménye egymás mellett
- ✅ Több aspektus: Training loss + validation loss + accuracy
- ✅ Idősorok: Különböző időperiódusok
- ✅ Exploratory Data Analysis (EDA): Gyors áttekintés

**💡 Tipp:** `figsize=(20, 10)` → első szám a szélesség, második a magasság!

### 📐 Matematikai Háttér: Sigmoid Függvény

A **szigmoid (sigmoid)** függvény az egyik legfontosabb aktivációs függvény a gépi tanulásban!

**Matematikai képlet:**
$$\sigma(x) = \frac{1}{1 + e^{-x}}$$

**Jellemzők:**
- 📈 **Kimenet tartomány:** (0, 1) - ezért ideális valószínűségekhez
- 🔄 **Alak:** S-alakú ("S-curve") görbület
- ⚖️ **Szimmetria:** σ(-x) = 1 - σ(x)
- 🎯 **Inflexiós pont:** x=0 helyen, érték=0.5

**Gépi tanulási alkalmazások:**
- 🧠 **Neurális hálók**: Aktivációs függvény rejtett rétegekben
- 📊 **Logisztikus regresszió**: Bináris osztályozás (0 vagy 1)
- 🎲 **Valószínűség becslés**: Kimenet = P(y=1|x)

**Derivált:** σ'(x) = σ(x) · (1 - σ(x)) - egyszerű számolni, ezért népszerű!

### 📊 Histogram Bins - Hogyan válaszd meg?

A **bins** paraméter meghatározza, hány "oszlopra" (kosárra) bontjuk az adatokat.

**Szintaxis:**
```python
plt.hist(data, bins=30, alpha=0.7)
```

**Paraméterek:**
- `bins`: Oszlopok száma (int) vagy élek listája (list)
- `alpha`: Átlátszóság (0=teljesen átlátszó, 1=teljesen átlátszatlan)
- `edgecolor`: Oszlopok szegélyének színe
- `density`: Ha True, normalizált (terület=1)

**Bins választása - Általános szabályok:**

```
KEVÉS adat (< 100):     bins = 10-20
KÖZEPES adat (100-1000): bins = 20-50
SOK adat (> 1000):      bins = 50-100
```

**Matematikai szabályok:**
1. **Sturges' formula**: bins = ⌈log₂(n) + 1⌉
2. **Square-root choice**: bins = ⌈√n⌉
3. **Rice Rule**: bins = ⌈2 · ∛n⌉

```python

bins_sturges = int(np.log2(n) + 1)  
# A Sturges-szabály alapján számolja ki a hisztogram osztályközeinek számát.
# Képlet: log2(n) + 1

bins_sqrt = int(np.sqrt(n))
```

### 📊 Normális Eloszlás Részletesen

A **normális (Gauss) eloszlás** a statisztika legfontosabb eloszlása!

**Matematikai képlet:**
$$f(x) = \frac{1}{\sigma\sqrt{2\pi}} e^{-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2}$$

Ahol:
- μ (mű): **átlag** (mean) - a görbe központja
- σ (szigma): **szórás** (standard deviation) - a görbe szélessége
- σ² : **variancia** - a szórás négyzete

**Paraméterek hatása:**
```
Átlag (μ):     Eltolás vízszintesen
  μ=0  →  Középen
  μ=5  →  Jobbra tolva
  μ=-3 →  Balra tolva

Szórás (σ):    Szélesség
  σ=0.5 →  Keskeny csúcs (adatok közel az átlaghoz)
  σ=1.0 →  Standard
  σ=2.0 →  Széles, lapos (adatok szétszórtak)
```

**68-95-99.7 szabály:**
```
μ-σ        μ        μ+σ        (68% az adatoknak)
  |--------|--------|
  
μ-2σ               μ+2σ       (95% az adatoknak)
  |----------------|

μ-3σ                      μ+3σ (99.7% az adatoknak)
  |----------------------|
```

**Gyakorlati példák:**
- 📏 **Testmagasság**: μ=170cm, σ=10cm
- 🧠 **IQ**: μ=100, σ=15
- 📊 **Hőmérséklet**: μ=20°C, σ=5°C

**Gépi tanulásban:**
- **Feature scaling**: Standardizálás → μ=0, σ=1
- **Súly inicializálás**: Xavier/He inicializálás normális eloszlással
- **Generatív modellek**: VAE, GAN - normális eloszlású látens tér

### 📦 Boxplot (Dobozábra) Értelmezése

A **boxplot** az egyik leghatékonyabb eszköz az adateloszlás gyors megértéséhez!

**Anatómia:**
```
        *  ← outlier (kiugró érték)
        │
        ┬  ← maximum (vagy 1.5×IQR)
        │
    ┌───┴───┐
    │       │  ← Q3 (75. percentilis)
    │   ━━  │  ← Medián (50. percentilis)
    │       │  ← Q1 (25. percentilis)
    └───┬───┘
        │
        ┴  ← minimum (vagy 1.5×IQR)
        │
        *  ← outlier
```

**Mit mutat?**
- **Doboz (box):** Az adatok középső 50%-a (IQR = Interquartile Range)
- **Vonal a dobozban:** Medián (középérték, ahol 50% alatta, 50% felette)
- **Bajuszok (whiskers):** Min/max (vagy 1.5×IQR távolság)
- **Pontok:** Outliers (kiugró értékek)

**Mire jó?**
1. ✅ **Gyors összehasonlítás:** Több csoport összehasonlítása egyszerre
2. ✅ **Outlier detektálás:** Látod a kiugró értékeket
3. ✅ **Szórás mérése:** Széles doboz = nagy variancia
4. ✅ **Szimmetria:** Medián a doboz közepén = szimmetrikus eloszlás

**IQR (Interquartile Range) jelentősége:**
```
IQR = Q3 - Q1
```
- Robusztus mérőszám (nem érzékeny outlierekre)
- Outlier definíció: Q1 - 1.5×IQR alatt VAGY Q3 + 1.5×IQR felett

**Példa értelmezés:**
- Keskeny doboz → konzisztens adatok, kis szórás
- Széles doboz → változatos adatok, nagy szórás
- Medián felül → balra ferde eloszlás
- Sok outlier → érdemes megvizsgálni miért!

### 🎯 loc vs iloc - A Legfontosabb Különbség

Ez a Pandas egyik legzavaróbb része kezdőknek - de egyszer megértve nagyon logikus!

| Szempont | `.loc[]` | `.iloc[]` |
|----------|----------|-----------|
| **Jelentés** | **loc**ation (címke szerint) | **i**nteger **loc**ation (pozíció szerint) |
| **Indexelés** | Címkék (label-based) | Egész számok (integer-based) |
| **Példa (sor)** | `df.loc['Hungary']` | `df.iloc[5]` |
| **Példa (oszlop)** | `df.loc[:, 'Age']` | `df.iloc[:, 2]` |
| **Szeletelés vége** | **Inkluzív** (tartalmazza) | **Exkluzív** (nem tartalmazza) |
| **Használat** | Ha ismered a címkét | Ha ismered a pozíciót |

**Szeletelési különbség példa:**
```python
df.loc[1:3]     # 1, 2, 3 indexű sorok (MIND A HÁROM!)
df.iloc[1:3]    # 1, 2 indexű sorok (3 NINCS benne, mint Python list)
```

**Amikor használd őket:**
- ✅ `.loc[]`: Amikor tudod az oszlop/sor nevét → `df.loc['Hungary', 'Population']`
- ✅ `.iloc[]`: Amikor pozíció alapján dolgozol → `df.iloc[0, 0]` (első sor, első oszlop)
- ✅ `.loc[]` boolean indexeléshez: `df.loc[df['Age'] > 30]`

**💡 Tipp:** Ha bizonytalan vagy, használj `.loc[]`-t - általában biztonságosabb!

#Korrelációs mátrix ábrázolása
A korreláció azt mutatja meg, hogy két változó között milyen összefüggés van a változások között. Ha két változó ugyanabban az irányban változik, akkor pozitív korrelációt mutatnak. Ha együttesen ellentétes irányban változnak (az egyik felfelé, a másik lefelé megy), akkor negatívan korrelálnak.

Kiszámíthatja az egyes attribútumpárok közötti korrelációt. Ezt korrelációs mátrixnak nevezzük. Ezután ábrázolhatja a korrelációs mátrixot, és képet kaphat arról, hogy mely változók korrelációja magas egymással.

Ezt azért hasznos tudni, mert egyes gépi tanulási algoritmusok, például a lineáris és logisztikus regresszió gyenge teljesítményt nyújthatnak, ha az adatokban magasan korrelált bemeneti változók vannak.

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ------------------------------------------------------------
# 0) Betöltés (a te kezdő kódoddal)
# ------------------------------------------------------------
df = pd.read_csv('https://raw.githubusercontent.com/karsarobert/MachineLearning2026/main/diabetes.csv')

print("Eredeti shape:", df.shape)
print("Eredeti oszlopok:", list(df.columns))

# ------------------------------------------------------------
# 1) Oszlopnevek "takarítása" (szóközök, kisbetűsítés)
#    - ez rengeteg KeyError-t megelőz
# ------------------------------------------------------------
df = df.copy()
df.columns = df.columns.str.strip().str.lower()

print("\nTisztított oszlopok:", list(df.columns))

# ------------------------------------------------------------
# 2) Elemszám ellenőrzés + hiányzók kezelése
# ------------------------------------------------------------
print("\nOszloponkénti elemszám (count):")
print(df.count())

print("\nHiányzó értékek (NaN) oszloponként:")
print(df.isnull().sum())

# Ha vannak NaN-ek: töltsük ki numerikus oszlopátlaggal
if df.isnull().sum().sum() > 0:
    num_cols = df.select_dtypes(include=[np.number]).columns
    df[num_cols] = df[num_cols].fillna(df[num_cols].mean())
    print("\nNaN-ek kitöltve numerikus oszlopátlaggal.")
    print("NaN ellenőrzés:")
    print(df.isnull().sum())

# ------------------------------------------------------------
# 3) Label oszlop megkeresése (Outcome / label / class / target stb.)
# ------------------------------------------------------------
label_candidates = ["outcome", "label", "target", "class", "y"]
label_col = None
for c in label_candidates:
    if c in df.columns:
        label_col = c
        break

if label_col is None:
    raise KeyError("Nem találok címke (label) oszlopot. Nézd meg: print(df.columns)")

print("\nLabel oszlop:", label_col)
print("Label egyedi értékek:", df[label_col].unique())

# ------------------------------------------------------------
# 4) Átlagok számítása címkénként (csak numerikus oszlopokra)
# ------------------------------------------------------------
num_cols = df.select_dtypes(include=[np.number]).columns
# (a label oszlop is numerikus lehet; a groupby-hoz kell, de az átlagból kivesszük)
feature_cols = [c for c in num_cols if c != label_col]

means_by_label = df.groupby(label_col)[feature_cols].mean()

print("\nÁtlagok címkénként (label=0/1):")
print(means_by_label)

# ------------------------------------------------------------
# 5) Pregnancies oszlop megkeresése (több lehetséges névvel)
# ------------------------------------------------------------
preg_candidates = [
    "pregnancies", "pregnancy", "pregnant", "n_pregnancies", "num_pregnancies",
    "preg", "preg_count", "pregn"
]

preg_col = None
for c in preg_candidates:
    if c in df.columns:
        preg_col = c
        break

# Ha nincs pontos találat: próbáljunk részszavas találatot
if preg_col is None:
    for col in df.columns:
        if "preg" in col:   # pl. "number_of_pregnancies"
            preg_col = col
            break

if preg_col is None:
    raise KeyError("Nem találok 'Pregnancies' jellegű oszlopot. Nézd meg: print(df.columns)")

print("\nPregnancies oszlop:", preg_col)

# ------------------------------------------------------------
# 6) Szűrés: ahol a szülések száma < 10
# ------------------------------------------------------------
df_filtered = df[df[preg_col] < 10].copy()
print("Szűrt shape (preg < 10):", df_filtered.shape)

# ------------------------------------------------------------
# 7) Korrelációs mátrix (szűrt adaton, numerikus oszlopok)
# ------------------------------------------------------------
num_cols_f = df_filtered.select_dtypes(include=[np.number]).columns
corr = df_filtered[num_cols_f].corr()

plt.figure(figsize=(8, 6))
plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45, ha="right")
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Korrelációs mátrix (preg < 10)")
plt.tight_layout()
plt.show()

# ------------------------------------------------------------
# 8) Hisztogramok (szűrt adaton)
# ------------------------------------------------------------
df_filtered[num_cols_f].hist(figsize=(12, 8))
plt.suptitle("Hisztogramok (preg < 10)")
plt.tight_layout()
plt.show()

# ------------------------------------------------------------
# 9)
# ------------------------------------------------------------
print("\nEloszlás tipp (ferdeség/skew alapján, csak tájékoztató):")
sk = df_filtered[num_cols_f].skew(numeric_only=True)
for col in sk.index:
    if abs(sk[col]) < 0.5:
        tip = "kb. szimmetrikus / közel normális"
    elif sk[col] > 0:
        tip = "jobbra ferde (right-skewed)"
    else:
        tip = "balra ferde (left-skewed)"
    print(f"- {col}: {tip} (skew={sk[col]:.2f})")
````

### 🔍 Mit Látunk a Hisztogramon?

**Amit meg kell értened:**

1. **Eloszlás alakja:**
   - 🔔 **Normális eloszlás** (harang alakú): Szimmetrikus, legtöbb adat a középen
   - 📈 **Jobbra ferde** (right-skewed): Hosszú farok jobbra → kevés nagy érték
   - 📉 **Balra ferde** (left-skewed): Hosszú farok balra → kevés kicsi érték

2. **Hosszú farok (Long tail):**
   - **Mit jelent?** Van néhány extrém nagy érték (outliers)
   - **Példák:** Jövedelmek (sok kicsi, kevés milliárdos), népesség (sok kis falu, kevés nagyváros)
   - **Probléma ML-ben:** Az átlag félrevezető! Inkább mediánt használj.

3. **Logaritmikus transzformáció (`np.log`):**
   - **Miért?** A hosszú farkú eloszlásokat "normálisabbá" teszi
   - **Hogyan?** Nagy értékeket összenyomja, kicsiket szétnyújtja
   - **Amikor használd:** Power-law eloszlások, exponenciális növekedés

```python
# Log:     [0, 1,  2,   3]    - kezelhető skála
```

### 🔗 Korrelációs Mátrix Értelmezése

A **korrelációs mátrix** megmutatja, hogy a változók mennyire mozognak együtt!

**Korreláció (Pearson-féle):**
- **Tartomány:** -1 és +1 között
- **+1:** Tökéletes pozitív korreláció (együtt nőnek)
- **0:** Nincs lineáris kapcsolat
- **-1:** Tökéletes negatív korreláció (az egyik nő, a másik csökken)

**Értelmezési skála:**
```
 0.0 - 0.2  →  Nagyon gyenge vagy nincs kapcsolat
 0.2 - 0.4  →  Gyenge kapcsolat
 0.4 - 0.6  →  Közepes kapcsolat
 0.6 - 0.8  →  Erős kapcsolat
 0.8 - 1.0  →  Nagyon erős kapcsolat
```

**Vizuális segítség (heatmap színek):**
- 🔴 **Piros/Sötét:** Erős pozitív korreláció
- ⚪ **Fehér/Világos:** Nincs korreláció
- 🔵 **Kék:** Erős negatív korreláció

**Mit keress a mátrixban?**
1. ✅ **Átló:** Mindig 1.0 (minden változó tökéletesen korrelál önmagával)
2. 🔍 **Erős korrelációk:** > 0.7 vagy < -0.7 → érdekes kapcsolatok!
3. ⚠️ **Multikollinearitás:** Ha két input változó erősen korrelál (>0.9), az lehet probléma ML-ben

**Gyakori félreértések:**
- ❌ **Korreláció ≠ okozat:** A fagylalteladás és a fuldoklások korrelálnak, de egyik sem okozza a másikat (mindkettőt a meleg időjárás okozza)
- ❌ **Csak lineáris kapcsolat:** Lehet erős nemlineáris kapcsolat is, amit a Pearson nem lát

**Gépi tanulásban:**
- **Feature selection:** Gyengén korreláló features-t elhagyhatod
- **Multikollinearitás:** Erősen korreláló input változók közül egyet törölj
- **Target korreláció:** Nézd meg, mely változók korrelálnak legjobban a célváltozóval!
